# Skin Cancer Detection using CNN

**Problem Statement**: To build a CNN-based model which can accurately detect melanoma. Melanoma is a type of cancer that can be deadly if not detected early. It accounts for 75% of skin cancer deaths. A solution that can evaluate images and alert dermatologists about the presence of melanoma has the potential to reduce a lot of manual effort needed in diagnosis.

***

## Import Necessary Libraries

```python
# Import necessary libraries
import pathlib
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
import PIL
import glob
import shutil
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
```

## Define Data Paths

```python
# Define the path for train and test images
# Adjust the paths according to your environment
data_dir_train = pathlib.Path('Train')  # Path to the training data
data_dir_test = pathlib.Path('Test')    # Path to the testing data

# Verify if the paths exist
print("Training directory exists:", data_dir_train.exists())
print("Testing directory exists:", data_dir_test.exists())
```

## Count Images in Train and Test Directories

```python
# Count the number of images in train and test directories
image_count_train = len(list(data_dir_train.glob('*/*.jpg')))
print("Total training images:", image_count_train)

image_count_test = len(list(data_dir_test.glob('*/*.jpg')))
print("Total testing images:", image_count_test)
```

## Define Parameters for Data Loading

```python
# Define batch size and image dimensions
batch_size = 32
img_height = 180
img_width = 180
```

## Create Training and Validation Datasets

```python
# Create training dataset
train_ds = tf.keras.preprocessing.image_dataset_from_directory(
    data_dir_train,
    validation_split=0.2,       # Use 20% of data for validation
    subset='training',          # Set as training data
    seed=123,                   # Set seed for reproducibility
    image_size=(img_height, img_width),
    batch_size=batch_size
)

# Create validation dataset
val_ds = tf.keras.preprocessing.image_dataset_from_directory(
    data_dir_train,
    validation_split=0.2,       # Use 20% of data for validation
    subset='validation',        # Set as validation data
    seed=123,                   # Set seed for reproducibility
    image_size=(img_height, img_width),
    batch_size=batch_size
)
```

## Get Class Names

```python
# Get class names
class_names = train_ds.class_names
print("Class names:", class_names)
```

## Visualize One Instance of Each Class

```python
# Visualize one instance of each class
plt.figure(figsize=(12, 12))
for i, class_name in enumerate(class_names):
    # Get the first image from the dataset corresponding to the current class
    for images, labels in train_ds:
        index = tf.where(labels == i)
        if len(index) > 0:
            image = images[index[0][0]]
            plt.subplot(3, 3, i + 1)
            plt.imshow(image.numpy().astype("uint8"))
            plt.title(class_name)
            plt.axis("off")
            break
plt.tight_layout()
plt.show()
```

## Configure Dataset for Performance

```python
# Cache and prefetch datasets for performance
AUTOTUNE = tf.data.AUTOTUNE

train_ds = train_ds.cache().shuffle(1000).prefetch(buffer_size=AUTOTUNE)
val_ds = val_ds.cache().prefetch(buffer_size=AUTOTUNE)
```

## Build the Initial CNN Model

```python
# Number of classes
num_classes = len(class_names)

# Create the initial model without data augmentation
model = Sequential([
    layers.Rescaling(1./255, input_shape=(img_height, img_width, 3)),  # Normalize pixel values
    layers.Conv2D(32, 3, padding='same', activation='relu'),           # First convolutional layer
    layers.MaxPooling2D(),                                             # First pooling layer
    layers.Conv2D(64, 3, padding='same', activation='relu'),           # Second convolutional layer
    layers.MaxPooling2D(),                                             # Second pooling layer
    layers.Conv2D(128, 3, padding='same', activation='relu'),          # Third convolutional layer
    layers.MaxPooling2D(),                                             # Third pooling layer
    layers.Flatten(),                                                  # Flatten layer
    layers.Dense(128, activation='relu'),                              # Fully connected layer
    layers.Dense(num_classes)                                          # Output layer
])
```

## Compile the Model

```python
# Compile the model with appropriate optimizer and loss function
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

## View the Model Summary

```python
# View the model summary
model.summary()
```

## Train the Initial Model

```python
# Train the model for 20 epochs
epochs = 20
history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=epochs
)
```

## Visualize Training Results

```python
# Retrieve training and validation accuracy and loss
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']

loss = history.history['loss']
val_loss = history.history['val_loss']

epochs_range = range(epochs)

# Plot training and validation accuracy and loss
plt.figure(figsize=(16, 8))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy (Initial Model)')

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss (Initial Model)')
plt.show()
```

## Analyze Findings After Initial Model Training

### Findings:
**Final Training Accuracy**: 88.24%

**Final Validation Accuracy**: 62.33%

**Observations:**

- The training accuracy reached approximately **88.24%**, while the validation accuracy reached **62.33%**.
- The training loss decreased steadily, indicating that the model is learning.
- The validation loss started to increase after epoch **overfitting**, while the validation accuracy plateaued or decreased, indicating overfitting.
- There's a gap between training and validation accuracy, suggesting that the model is overfitting to the training data.

## Apply Data Augmentation to Reduce Overfitting

```python
# Apply data augmentation to reduce overfitting
data_augmentation = keras.Sequential(
    [
        layers.RandomFlip("horizontal_and_vertical"),  # Randomly flip images
        layers.RandomRotation(0.2),                    # Randomly rotate images
        layers.RandomZoom(0.2),                        # Randomly zoom images
    ]
)
```

## Visualize Augmentation Strategy on One Image

```python
# Visualize the augmentation strategy on one image
for images, _ in train_ds.take(1):
    plt.figure(figsize=(10, 10))
    first_image = images[0]
    for i in range(9):
        augmented_image = data_augmentation(tf.expand_dims(first_image, 0))
        ax = plt.subplot(3, 3, i + 1)
        plt.imshow(augmented_image[0].numpy().astype("uint8"))
        plt.axis("off")
    plt.tight_layout()
    plt.show()
```

## Build the Model with Data Augmentation and Dropout

```python
# Create the model with data augmentation and dropout layers
model = Sequential([
    data_augmentation,                                  # Include data augmentation
    layers.Rescaling(1./255),                           # Normalize pixel values
    layers.Conv2D(32, 3, padding='same', activation='relu'),
    layers.MaxPooling2D(),
    layers.Dropout(0.2),                                # Add dropout layer to reduce overfitting
    layers.Conv2D(64, 3, padding='same', activation='relu'),
    layers.MaxPooling2D(),
    layers.Dropout(0.2),
    layers.Conv2D(128, 3, padding='same', activation='relu'),
    layers.MaxPooling2D(),
    layers.Dropout(0.2),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(num_classes)
])
```

## Compile the Model

```python
# Compile the model with appropriate optimizer and loss function
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

## Train the Model with Data Augmentation

```python
# Train the model for 20 epochs
epochs = 20
history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=epochs
)
```

## Visualize Training Results After Data Augmentation

```python
# Retrieve training and validation accuracy and loss
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']

loss = history.history['loss']
val_loss = history.history['val_loss']

epochs_range = range(epochs)

# Plot training and validation accuracy and loss
plt.figure(figsize=(16, 8))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy (After Data Augmentation)')

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss (After Data Augmentation)')
plt.show()
```

## Analyze Findings After Model Training with Data Augmentation

### Findings:
**Final Training Accuracy**: 53.19%

**Final Validation Accuracy**: 54.56%

Observations:

- After applying data augmentation and dropout, the validation accuracy improved to **54.56%**.
- The gap between training and validation accuracy reduced, indicating **less overfitting**.
- The validation loss decreased, showing that the model generalizes better to unseen data.
- Therefore, data augmentation and regularization techniques **helped mitigate overfitting**.

## Examine Class Distribution in Training Dataset

```python
# Examine the class distribution in the training dataset
train_dir = 'Train'

class_counts = {}
for class_name in class_names:
    class_dir = os.path.join(train_dir, class_name)
    # Count images in the class directory
    count = len(glob.glob(os.path.join(class_dir, '*.jpg')))
    class_counts[class_name] = count

# Display class counts
print("Class counts before augmentation:")
for class_name, count in class_counts.items():
    print(f"{class_name}: {count} images")

# Plot class distribution
plt.figure(figsize=(10, 6))
plt.bar(class_counts.keys(), class_counts.values())
plt.xticks(rotation=45)
plt.title('Class Distribution Before Augmentation')
plt.xlabel('Class Names')
plt.ylabel('Number of Samples')
plt.show()

# Identify classes with least and most samples
least_samples_class = min(class_counts, key=class_counts.get)
most_samples_class = max(class_counts, key=class_counts.get)
print(f"\nClass with least samples: {least_samples_class} ({class_counts[least_samples_class]} images)")
print(f"Class with most samples: {most_samples_class} ({class_counts[most_samples_class]} images)")
```

## Rectify Class Imbalances Using Augmentor

```python
# Install Augmentor library
!pip install Augmentor

# Import Augmentor
import Augmentor

# Rectify class imbalances using Augmentor
for class_name in class_names:
    class_dir = os.path.join(train_dir, class_name)
    p = Augmentor.Pipeline(class_dir)
    p.rotate(probability=0.7, max_left_rotation=10, max_right_rotation=10)
    p.flip_left_right(probability=0.5)
    p.zoom_random(probability=0.5, percentage_area=0.9)
    # Calculate number of samples to generate to reach 500 images per class
    current_count = class_counts[class_name]
    samples_to_generate = 500 - current_count if 500 > current_count else 0
    if samples_to_generate > 0:
        p.sample(samples_to_generate)
```

## Update Class Counts After Augmentation

```python
# Update class counts after augmentation
class_counts_after = {}
for class_name in class_names:
    class_dir = os.path.join(train_dir, class_name)
    # Count images in the class directory, including images in 'output' directory
    count = len(glob.glob(os.path.join(class_dir, '*.jpg'))) + len(glob.glob(os.path.join(class_dir, 'output', '*.jpg')))
    class_counts_after[class_name] = count

# Display class counts after augmentation
print("Class counts after augmentation:")
for class_name, count in class_counts_after.items():
    print(f"{class_name}: {count} images")

# Plot class distribution after augmentation
plt.figure(figsize=(10, 6))
plt.bar(class_counts_after.keys(), class_counts_after.values())
plt.xticks(rotation=45)
plt.title('Class Distribution After Augmentation')
plt.xlabel('Class Names')
plt.ylabel('Number of Samples')
plt.show()
```

## Create New Training and Validation Datasets Including Augmented Images

```python
# Create new training and validation datasets including augmented images

# Create training dataset including augmented images
train_ds = tf.keras.preprocessing.image_dataset_from_directory(
    data_dir_train,
    validation_split=0.2,
    subset='training',
    seed=123,
    image_size=(img_height, img_width),
    batch_size=batch_size,
    shuffle=True
)

# Create validation dataset
val_ds = tf.keras.preprocessing.image_dataset_from_directory(
    data_dir_train,
    validation_split=0.2,
    subset='validation',
    seed=123,
    image_size=(img_height, img_width),
    batch_size=batch_size,
    shuffle=True
)
```

## Configure Dataset for Performance

```python
# Cache and prefetch datasets
AUTOTUNE = tf.data.AUTOTUNE

train_ds = train_ds.cache().shuffle(1000).prefetch(buffer_size=AUTOTUNE)
val_ds = val_ds.cache().prefetch(buffer_size=AUTOTUNE)
```

## Build the Model on Balanced Data

```python
# Create the model (with normalization and dropout)
model = Sequential([
    layers.Rescaling(1./255, input_shape=(img_height, img_width, 3)),
    layers.Conv2D(32, 3, padding='same', activation='relu'),
    layers.MaxPooling2D(),
    layers.Dropout(0.3),
    layers.Conv2D(64, 3, padding='same', activation='relu'),
    layers.MaxPooling2D(),
    layers.Dropout(0.3),
    layers.Conv2D(128, 3, padding='same', activation='relu'),
    layers.MaxPooling2D(),
    layers.Dropout(0.3),
    layers.Flatten(),
    layers.Dense(256, activation='relu'),
    layers.Dense(num_classes)
])
```

## Compile the Model

```python
# Compile the model with appropriate optimizer and loss function
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

## Train the Model on Balanced Data

```python
# Train the model for 30 epochs
epochs = 30
history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=epochs
)
```

## Visualize Training Results After Handling Class Imbalance

```python
# Retrieve training and validation accuracy and loss
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']

loss = history.history['loss']
val_loss = history.history['val_loss']

epochs_range = range(epochs)

# Plot training and validation accuracy and loss
plt.figure(figsize=(16, 8))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy (After Handling Class Imbalance)')

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss (After Handling Class Imbalance)')
plt.show()
```

## Analyze Findings After Model Training on Balanced Data

### Findings:
**Final Training Accuracy**: 94.06%

**Final Validation Accuracy**: 73.96%

Observations:

- After handling class imbalance, the model's validation accuracy improved to 73.96%.
- The confusion matrix shows better performance across all classes, indicating that the model is no longer biased towards classes with more samples.
- The overfitting issue has been further reduced, as evidenced by closer training and validation metrics.
- Class rebalance helped in **improving model generalization**.

## Evaluate Model Performance Using Confusion Matrix

```python
# Install sklearn library
!pip install scikit-learn

# Evaluate model performance using confusion matrix and classification report
from sklearn.metrics import confusion_matrix, classification_report
import numpy as np

# Generate predictions on the validation dataset
y_pred = []
y_true = []

for images, labels in val_ds:
    predictions = model.predict(images)
    predicted_classes = np.argmax(predictions, axis=1)
    y_pred.extend(predicted_classes)
    y_true.extend(labels.numpy())

# Compute confusion matrix
cm = confusion_matrix(y_true, y_pred)
print('Confusion Matrix')
print(cm)

# Display classification report
print('Classification Report')
target_names = class_names
print(classification_report(y_true, y_pred, target_names=target_names))
```

***
## Analysis of the Confusion Matrix and Classification Report

### Confusion matrix and classification report indicate:

**Overall Accuracy**: 74% on the validation dataset.

**Class-wise Performance**:
- _High Precision and Recall_: For classes like dermatofibroma and vascular lesion, indicating the model performs well on these classes.
- _Moderate Performance_: For classes like actinic keratosis, seborrheic keratosis, and squamous cell carcinoma.
- _Lower Performance_: For melanoma, with precision and recall around 57% and 47%, respectively.
  
**Observations**:

- _Improved Class Balance_: After handling class imbalance, the model's performance across classes has improved.
- _Reduced Overfitting_: The gap between training and validation accuracy has decreased, indicating reduced overfitting.
- _Model Generalization_: The model generalizes better to unseen data after addressing class imbalance.
